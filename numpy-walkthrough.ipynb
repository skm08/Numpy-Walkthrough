**class numpy.ndarray(shape, dtype=float, buffer=None, offset=0, strides=None, order=None)**

```python
# If buffer is None, then only shape, dtype, and order are used
np.ndarray(shape=(2,2),dtype=float,order='F')
```

```python
# If buffer is an object exposing the buffer interface, then all keywords are interpreted.

np.ndarray((2,),buffer=np.array([1,2,3]),offset=np.int_().itemsize,dtype=int)
```

**numpy.reshape**

```python
a=np.arange(6)
```

```python
a
```

```python
a=a.reshape((3,2))
```

```python
a
```

```python
np.reshape(a,(2,3)) # C-like index ordering
```

```python
np.reshape(np.ravel(a),(2,3)) # equivalent to C ravel then C reshape
```

```python
np.reshape(a, (2, 3), order='F') # Fortran-like index ordering
```

```python
np.reshape(np.ravel(a, order='F'), (2, 3), order='F')
```

```python
a = np.array([[1,2,3], [4,5,6]])
```

```python
a
```

```python
np.reshape(a, 6)
```

```python
np.reshape(a, 6, order='F')
```

```python
np.reshape(a, (3,-1))       # the unspecified value is inferred to be 2
```
