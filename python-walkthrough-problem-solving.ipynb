**Write a function that returns the lesser of two given numbers if both numbers are even, but returns the greater if one or both numbers are odd**

```python
def two_evens(x,y):
    if x%2==0 and y%2==0:
        return min(x,y)
    else:
        return  max(x,y)
```

```python
two_evens(2,4),two_evens(3,5)
```

```python
def Capitalizer(name):
    first_part=name[:3]
    second_part=name[3:]
    return first_part.capitalize() + second_part.capitalize()
```

**Given a list of ints, return True if the array contains a 3 next to a 3 somewhere.**

```python
def three_next(x):
    for p in range(0,len(x)-1):
        if (x[p]==3 and x[p+1]==3):
            return True
    return False
```

```python
three_next([1,2,3,3])
```

**Given three integers between 1 and 11, if their sum is less than or equal to 21, return their sum. If their sum exceeds 21 and there's an eleven, reduce the total sum by 10. Finally, if the sum (even after adjustment) exceeds 21, return 'BUST'**

```python
def sum11(a,b,c):
    if sum((a,b,c))<=21:
        return sum((a,b,c))
    elif sum((a,b,c))>21 and 11 in (a,b,c):
        return sum((a,b,c)) - 10
    else:
        return 'BUST'
```

```python
sum11(2,5,9)
```

**Return the sum of the numbers in the array, except ignore sections of numbers starting with a 6 and extending to the next 9 (every 6 will be followed by at least one 9). Return 0 for no numbers.**

```python
def sixty_nine(arr):
    total = 0
    add = True
    for num in arr:
        while add:
            if num != 6:
                total += num
                break
            else:
                add = False
        while not add:
            if num != 9:
                break
            else:
                add = True
                break
    return total
```

```python
sixty_nine([2, 1, 6, 9, 11])
```

**Write a function that takes in a list of integers and returns True if it contains 007 in order**

```python
def zeroseven(nums):

    code = [0,0,7]
    
    for num in nums:
        if num == code[0]:
            code.pop(0)
       
    return len(code) == 0
```

```python
zeroseven([1,0,2,4,0,5,7])
```

**Write a function that returns the number of prime numbers that exist up to and including a given number**

```python
def primes(num):
    primes=[2,3]
    x=5
    while x<=num:
        for y in range(3,x,2):
            if x%y == 0:
                x += 2
                break
        else:
            primes.append(x)
            x += 2
    print(primes)
    return len(primes)
```

```python
primes(100)
```

**Write a function that checks whether a number is in a given range (inclusive of high and low)**

```python
def ran_check(num,low,high):
    for x in range(low,high+1):
        if x == num:
            return True
    
    return False
```

**Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters.**

```python
def up_low(s):
    lowercase=0
    uppercase=0
    
    for char in s:
        if char.isupper():
            uppercase +=1
        elif char.islower():
            lowercase += 1
        else:
            pass
    print(f"Sample String : {s}")
    print(f"No. of Upper case characters : {uppercase}")
    print(f"No. of lower case characters : {lowercase}")
```

```python
s = 'Hello Mr. Rogers, how are you this fine Tuesday?'
up_low(s)
```

**Write a Python function that checks whether a word or phrase is palindrome or not.**

```python
def palindrome(s):
    s=s.replace(' ','')
    return s == s[::-1]
```

**Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)**

```python
import string

def ispangram(str1, alphabet=string.ascii_lowercase):
    alphaset=set(alphabet)
    str1=str1.replace(' ','')
    str1=str1.lower()
    str1=set(str1)
    return str1 == alphaset
```

```python
ispangram("The quick brown fox jumps over the lazy dog")
```
